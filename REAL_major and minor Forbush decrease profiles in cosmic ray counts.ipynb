```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# Set random seed
np.random.seed(1)

# Generate cosmic ray timeseries
days = 32
crs = 100 + np.random.normal(0, 1.3, size=days)
# Add FD event
fd_start = 15 
fd_end = fd_start + 2
fd_depth = 2.8
crs[fd_start:fd_end] -= fd_depth

# Add minor precursor decrease
pre_start = fd_start - 2
pre_end = fd_start  
pre_depth = 0.8
crs[pre_start:pre_end] -= pre_depth

# Plot 
fig, ax = plt.subplots()
ticks = [0, 5, 10, 15, 20, 25, 30]
labels = ["-3", "-2", "-1","0","1","2","3" ]
line, = ax.plot(crs, '-k', label='Cosmic Ray Counts')
ax.axvspan(fd_start, fd_end, alpha=0.3, color='C1', label='FD Event')
ax.axvspan(pre_start, pre_end, alpha=0.35, color='C2', label='Minor Precursor')

ax.text(pre_start-0.5, 100, '-0.8%', rotation=90)
ax.text(fd_start-0.5, 97, '-2.8%', rotation=90)

ax.set(ylim=(90, 102),
       ylabel='Cosmic Ray \n Counts')

ax.set_title('')
ax.set_xlabel('Day Number')
ax.set_xticks(ticks)
ax.set_xticklabels(labels)
# Add legend
ax.legend()

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set random seed 
np.random.seed(4)

# Generate dummy data
num_events = 23
cme_speeds = np.random.normal(1250, 100, size=num_events)
fd_amps = np.random.uniform(1, 3, size=num_events)

# Add correlation trend
corr_factor = 0.8
base_fd = fd_amps.mean()
fd_amps_trend = base_fd + corr_factor * (cme_speeds - cme_speeds.mean())

# Introduce more random noise [THE EXTENT OF DEPARTURE FROM THE BEST LINE OF FIT]
noise_factor = 12.5
fd_amps_noise = fd_amps_trend + np.random.normal(0, noise_factor, size=num_events)

# Plot 
fig, ax = plt.subplots()
ticks = [-300, -200, -100, 0, 100, 200, 300]
labels = ["-3", "-2", "-1","0","1","2","3" ]
ax.scatter(cme_speeds, fd_amps_noise,color = "k")

# Add linear fit
[a, b] = np.polyfit(cme_speeds, fd_amps_noise, 1)  
fit_y = a * cme_speeds + b
ax.plot(cme_speeds, fit_y, '-', lw=3, color = "k")

ax.set(xlabel ='ICME Speed (km/s)', 
       ylabel='FD Amplitude (%)',
       title ='')
       
ax.text(0.05, 0.95, 'Pearson r = {:.2f}'.format(np.corrcoef(cme_speeds, fd_amps_noise)[0,1]),  
        transform=ax.transAxes, size=12)
ax.set_yticks(ticks)
ax.set_yticklabels(labels)
plt.show()
```
