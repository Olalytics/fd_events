```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# Set random seed
np.random.seed(1)

# Generate cosmic ray timeseries
days = 32
crs = 100 + np.random.normal(0, 1.3, size=days)
# Add FD event
fd_start = 15 
fd_end = fd_start + 2
fd_depth = 2.8
crs[fd_start:fd_end] -= fd_depth

# Add minor precursor decrease
pre_start = fd_start - 2
pre_end = fd_start  
pre_depth = 0.8
crs[pre_start:pre_end] -= pre_depth

# Plot 
fig, ax = plt.subplots()
ticks = [0, 5, 10, 15, 20, 25, 30]
labels = ["-3", "-2", "-1","0","1","2","3" ]
line, = ax.plot(crs, '-k', label='Cosmic Ray Counts')
ax.axvspan(fd_start, fd_end, alpha=0.3, color='C1', label='FD Event')
ax.axvspan(pre_start, pre_end, alpha=0.35, color='C2', label='Minor Precursor')

ax.text(pre_start-0.5, 100, '-0.8%', rotation=90)
ax.text(fd_start-0.5, 97, '-2.8%', rotation=90)

ax.set(ylim=(90, 102),
       ylabel='Cosmic Ray \n Counts')

ax.set_title('')
ax.set_xlabel('Day Number')
ax.set_xticks(ticks)
ax.set_xticklabels(labels)
# Add legend
ax.legend()

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Function to generate ellipse points
def generate_ellipse_points(a, b, tilt_angle, num_points=100):
    theta = np.linspace(0, 2 * np.pi, num_points)
    x = a * np.cos(theta)
    y = b * np.sin(theta)
    z = 0  # Assume a 2D ellipse in the xy-plane
    # Rotate the ellipse to match the tilt angle
    x_rot = x * np.cos(tilt_angle) - y * np.sin(tilt_angle)
    y_rot = x * np.sin(tilt_angle) + y * np.cos(tilt_angle)
    z_rot = z
    return x_rot, y_rot, z_rot

# Function to plot the shock model
def plot_shock_model():
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Shock model parameters
    a = 1.5  # Semi-major axis
    b = 0.5  # Semi-minor axis
    tilt_angle = np.radians(30)  # Tilt angle in radians

    # Generate ellipse points
    x, y, z = generate_ellipse_points(a, b, tilt_angle)

    # Plot ellipse
    ax.plot(x, y, z, label='Shock Front', color='b')

    # Add vector arrows for density and magnetic field directionality
    arrow_length = 1.5
    ax.quiver(0, 0, 0, arrow_length, 0, 0, color='r', label='Density Direction')
    ax.quiver(0, 0, 0, 0, arrow_length, 0, color='g', label='Magnetic Field Direction')

    # Set plot labels
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    #ax.set_title('3D Schematic of Inclined Ellipse-Shaped Shock Front')
    ax.legend()

    plt.show()

# Plot the shock model
plot_shock_model()
```

```python
import numpy as np 
from scipy import stats
import matplotlib.pyplot as plt

# Set random seed 
np.random.seed(3)


num_cmes = 23  
cme_times = np.cumsum(np.random.randint(1, 4, num_cmes))
cme_speeds = np.random.normal(1250, 150, num_cmes)

# Cosmic ray data with FD signal    
cr_days = 50
cr_counts = 100 + np.random.normal(0, 2.3, cr_days)

for cme_t in cme_times:
    
    fd_start = cme_t
    fd_end = fd_start + np.random.randint(0, 2)  
    fd_delta = np.random.uniform(-4, -0.8) 
    cr_counts[fd_start:fd_end] += fd_delta

# Correlation analysis
fd_amps = np.subtract(cr_counts[cme_times-1], cr_counts[cme_times])
rho, pval = stats.spearmanr(cme_speeds, -fd_amps)

print(f'CME-FD Correlation: rho={rho:.2f}, pval={pval:.3f}')

# Plot CR & FD signals 
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 4), sharey=True)
ax1.plot(cr_counts)
ax1.scatter(cme_times, cr_counts[cme_times], marker='v', c='r')  

# Set x-axis ticks and labels for the left subplot (ax1)
ax1.set_xticks(cme_times)
ax1.set_xticklabels(range(1, num_cmes + 1))
ax1.set_xlabel('CME Event')

ax2.scatter(cme_speeds, -fd_amps)
ax2.set(xlabel='CME Speed (km/s)', ylabel='FD Amplitude (%)')  

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up parameters
num_frames = 100
initial_width = 90.0  # Initial angular width in degrees
propagation_distance = 570.0  # Propagation distance in solar radii
expansion_factor = 30  # Adjust as needed

# Function to calculate shock geometry at each frame
def calculate_shock_geometry(frame, initial_width, propagation_distance, expansion_factor):
    angular_width = initial_width / (1 + expansion_factor * frame / num_frames)  # Narrower toward Earth
    distance_from_sun = propagation_distance * np.cos(np.radians(angular_width / 2))
    return angular_width, distance_from_sun

# Generate frames
frames = []
distances = []
for frame in range(num_frames):
    angular_width, distance_from_sun = calculate_shock_geometry(frame, initial_width, propagation_distance, expansion_factor)
    frames.append(angular_width)
    distances.append(distance_from_sun)

# Generate difference imagery
cme_early_image = np.zeros((1024, 1024))  # Placeholder for early image
cme_late_image = np.zeros((1024, 1024))  # Placeholder for late image

# Calculate base difference image
diff_image = cme_late_image - cme_early_image

# Plot angular spans
fig, ax = plt.subplots()
ax.plot(frames, distances, marker='o', linestyle='-', color='blue')
ax.set(xlabel='Angular Width (degrees)', ylabel='Distance from Sun (solar radii)')

plt.show()
```
